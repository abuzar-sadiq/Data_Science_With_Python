# Grid Plot

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# import tips that already done
tips.head()
```

```python
g=sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
g=g.map(plt.hist,'total_bill')
```

```python
g=sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
g=g.map(plt.scatter,'total_bill','tip').add_legend()
```
